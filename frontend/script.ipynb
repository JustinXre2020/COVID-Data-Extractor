```python
# generate two tables
import pandas as pd
import random
import os
import names
import json
```

```python
with open(os.path.join(os.getcwd(), 'location.txt'), 'r') as f:
    regions = f.readlines()
regions = [region.strip() for region in regions]
```

```python
patientRegion = []
patientName = []
patientId = []
fullyVaccinated = []
haveBooster = []
id = 1

for i in range(1500):
    patientName.append(names.get_full_name())
    patientId.append(f"{id:04n}")
    patientRegion.append(random.choice(regions))
    haveTwoShoots = random.choice([True, False])
    fullyVaccinated.append(haveTwoShoots)
    if haveTwoShoots:
        haveBooster.append(random.choice([True, False]))
    else:
        haveBooster.append(False)
    id += 1
```

```python
df_patients = pd.DataFrame({"PatientName": patientName, "PatientId": patientId, "Region": patientRegion})
df_patients
```

```python
df_epidemiology = pd.read_csv("C:/Users/justi/Downloads/epidemiology.csv")
df_epidemiology
```

```python
df_epidemiology['Region'] = df_epidemiology['Region'].apply(lambda x: x[:2])
df_epidemiology
```

```python
lll = list(set([i[:2] for i in df_epidemiology['Region'] if i[:2] in regions]))
print(len(lll))
```

```python
df_epidemiology['date'] = df_epidemiology['date'].apply(lambda x : pd.to_datetime(x))
df_epidemiology
```

```python
df_epidemiology_ = df_epidemiology[(df_epidemiology['date'] == '2022-09-23') & (df_epidemiology['Region'] == 'AU')]
print(f"The sum is: {df_epidemiology_['cumulative_tested'].sum()}")
```

```python
df_epidemiology_group = df_epidemiology.groupby(['Region', pd.Grouper(key='date', axis=0, freq='1D')]).agg('sum').reset_index()
df_epidemiology_group
```

```python
df_epidemiology_group['cumulative_recovered'] = [i if i != 0 else None for i in df_epidemiology_group['cumulative_recovered']]
df_epidemiology_group['cumulative_tested'] = [i if i != 0 else None for i in df_epidemiology_group['cumulative_tested']]
df_epidemiology_group
```

```python
df_epidemiology_group.to_csv('epidemiology.csv', index=False)
```
